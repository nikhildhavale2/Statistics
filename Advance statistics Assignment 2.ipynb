**Q2 What is a p-value, and how is it used in hypothesis testing? What does it mean if the p-value is very small (e.g., 0.01)?**

**P-Value: Definition and Role in Hypothesis Testing**

A **p-value** is a measure of the probability that an observed difference could have occurred just by random chance. In the context of hypothesis testing, it helps determine the strength of the evidence against the null hypothesis (H0).

**Interpreting the P-Value:**

**High P-Value:** A high p-value (typically > 0.05) indicates weak evidence against the null hypothesis, so you fail to reject it. This means that the observed data is consistent with the null hypothesis.

**Low P-Value:** A low p-value (typically ‚â§ 0.05) indicates strong evidence against the null hypothesis, so you reject it. This suggests that the observed data is unlikely to have occurred under the null hypothesis.

**Steps in Hypothesis Testing Using P-Value:**

**1. State the Hypotheses:**

**Null Hypothesis (H0):** There is no effect or difference. For example, ùúá=ùúá0.

**Alternative Hypothesis (H1):** There is an effect or difference. For example, ùúá‚â†ùúá0.

**2. Determine the Significance Level (Œ±):**

Common significance levels are 0.05, 0.01, or 0.10. This is the threshold for deciding whether to reject the null hypothesis.

**3. Collect Data and Calculate the Test Statistic:**

Calculate the appropriate test statistic (e.g., z-score, t-score) based on the data.

**4. Calculate the P-Value:**

Use the test statistic to find the p-value from the relevant statistical distribution (e.g., standard normal distribution).

**5. Compare P-Value to Significance Level:**

If ùëù‚â§ùõº, reject the null hypothesis.

If ùëù>ùõº, fail to reject the null hypothesis.

**Example Interpretation:**
**Very Small P-Value (e.g., 0.01):**

A p-value of 0.01 means there is only a 1% chance that the observed data would occur if the null hypothesis were true.

This provides strong evidence against the null hypothesis.

If the significance level is 0.05, you would reject the null hypothesis because 0.01 < 0.05.

**Practical Implications:**
Statistical Significance: A very small p-value indicates that the observed effect is statistically significant, suggesting that the effect is unlikely to be due to random variation alone.

Real-World Decisions: While a low p-value indicates statistical significance, it doesn‚Äôt imply practical significance. It‚Äôs important to consider the effect size and real-world relevance of the findings.

**Q6. Define the terms "probability distribution" and "probability density function" (PDF). How does a PDF differ from a probability mass function (PMF)**

**Probability Distribution**

A probability distribution describes how the values of a random variable are distributed. It provides the probabilities of occurrence of different possible outcomes in an experiment. Probability distributions can be discrete or continuous.

Probability Density Function (PDF)
A Probability Density Function (PDF) is used for continuous random variables. It represents the relative likelihood of a continuous random variable to take on a particular value. The PDF gives the probability that the random variable falls within a particular range of values, as opposed to taking on any one value.

**Characteristics:**

The value of the PDF at a given point represents the density of the probability at that point, not the actual probability.

The area under the PDF curve over an interval represents the probability of the variable falling within that interval.

The total area under the PDF curve over the entire range of possible values is equal to 1.

**Probability Mass Function (PMF)**

A Probability Mass Function (PMF) is used for discrete random variables. It gives the probability that a discrete random variable is exactly equal to some value.

**Characteristics:**

The PMF assigns probabilities to each possible value of the discrete random variable.

The sum of all probabilities in the PMF is equal to 1.

**Examples**

**PDF Example:**

Consider the height of individuals. The height can be any value within a range (e.g., 150 cm to 200 cm), and the PDF shows the density of probabilities for different height values.

**PMF Example:**

Consider the number of heads obtained when flipping a coin 10 times. The possible values are discrete (0 to 10), and the PMF gives the probability of each outcome (e.g., getting exactly 5 heads).

**Q.7 Explain the Central Limit Theorem (CLT) with example.**

**Central Limit Theorem (CLT)**

The Central Limit Theorem (CLT) is one of the most important concepts in statistics. It states that the sampling distribution of the sample mean (or sum) of a sufficiently large number of independent, identically distributed (i.i.d.) random variables, regardless of the shape of the original population distribution, will approximate a normal distribution. This convergence to the normal distribution occurs even if the original variable is not normally distributed.

**Key Points of the CLT:**

**Sample Size:** The theorem applies to sufficiently large sample sizes. Typically, a sample size greater than 30 is considered adequate, though the required size can vary depending on the original distribution.

**Independence:** The samples must be independent of each other.

**Identically Distributed:** The samples must come from the same distribution with a fixed mean and variance.

**Mathematical Expression:**

If ùëã1,ùëã2,‚Ä¶,ùëãùëõare i.i.d. random variables with mean ùúá and variance ùúé2, then the sample mean ùëã will be approximately normally distributed with mean ùúá and variance ùúé 2 ùëõ as ùëõ becomes large.

**Example:**

Let's say we want to understand the average height of adult women in a city. The population distribution of heights is not known to be normal, but we can use the CLT to approximate the distribution of the sample mean.

**1. Population Mean and Variance:** Assume the mean height (ùúá) is 165 cm, and the standard deviation (ùúé) is 10 cm.

**2. Sample Size:** We take a random sample of 50 women (n = 50).

According to the CLT, the distribution of the sample mean height will be approximately normal, even if the original height distribution is not normal.

**Calculation:**

Mean of the Sampling Distribution: The mean (ùúáùëãÀâ) of the sampling distribution will be the same as the population mean, which is 165 cm.

**Standard Error:** The standard deviation of the sampling distribution (standard error) is given by ùúé/‚àöùëõ.

Standard error = 10/‚àö50 ‚âà 1.41

Therefore, the sampling distribution of the sample mean height of 50 women will be approximately normally distributed with:

Mean (ùúáùëã) = 165 cm

Standard Error (ùúéùëã) = 1.41 cm

**Visualization:**

If we were to take multiple samples of 50 women and calculate the sample mean height for each sample, we would see that the distribution of these sample means forms a normal distribution centered around 165 cm with a standard error of 1.41 cm.

**Practical Implications:**

**Inference:** The CLT allows us to make inferences about population parameters using the normal distribution, even when the population distribution is unknown.

**Confidence Intervals:** It helps in constructing confidence intervals for population means.

**Hypothesis Testing:** Facilitates hypothesis testing for population means using z-scores and t-scores.

**Q9 Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis?**

**Task: Write Python code to calculate the z-score and p-value for the given data.**

**Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testing.**

Given Data:
Sample mean (ùë•) = 105

Population mean (ùúá) = 100

Standard deviation (ùúé) = 15

Sample size (ùëõ) = 25

Significance level (ùõº) = 0.05

**Calculate the P-value**

The p-value can be found using the cumulative distribution function (CDF) of the standard normal distribution. Since it's a two-tailed test, we need to consider both tails of the distribution.

Here is the Python code to calculate the z-score and p-value:

```python
import scipy.stats as stats
import math

# Given data
sample_mean = 105
population_mean = 100
std_dev = 15
sample_size = 25
significance_level = 0.05

# Calculate the z-score
z_score = (sample_mean - population_mean) / (std_dev / math.sqrt(sample_size))

# Calculate the p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Determine whether to reject or fail to reject the null hypothesis
reject_null = p_value < significance_level

# Output the results
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
print(f"Reject null hypothesis: {reject_null}")
```

**Interpretation:**

Z-score: A z-score of approximately 1.67 means that the sample mean is 1.67 standard deviations above the population mean.

P-value: This value tells us the probability of observing a sample mean as extreme as 105, assuming the null hypothesis is true.

**Decision:**
If the p-value is less than the significance level (0.05), we reject the null hypothesis.

If the p-value is greater than the significance level, we fail to reject the null hypothesis.

**Q.10 Simulate a binomial distribution with 10 trials and a probability of success of 0.6 using Python. Generate 1,000 samples and plot the distribution. What is the expected mean and variance?**
